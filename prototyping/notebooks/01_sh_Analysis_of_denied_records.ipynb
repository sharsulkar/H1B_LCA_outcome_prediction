```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
LCA_df=pd.read_excel('https://www.dol.gov/sites/dolgov/files/ETA/oflc/pdfs/LCA_Disclosure_Data_FY2020_Q2.xlsx')
```

```
LCA_df_copy=LCA_df.copy()
```

```
LCA_df.columns
```

```
#task list -
#1. Find the distribution of data vs outcomes
#2. Remove rows that are not Certified or Denied
#3. Checkout Denied rows to see whats going on there?
```

```
#1. Find the distribution of data vs outcomes
LCA_df.groupby('CASE_STATUS').size()
```

```
#2. Remove rows that are not Certified or Denied
LCA_df.drop(LCA_df[LCA_df.CASE_STATUS.isin(['Certified - Withdrawn', 'Withdrawn'])].index,inplace=True)
```

```
#3. Checkout Denied rows to see whats going on there?
LCA_df[LCA_df.CASE_STATUS.isin(['Denied'])]
```

```
#non-H-1B applications
LCA_df[LCA_df.VISA_CLASS!='H-1B'].groupby(['CASE_STATUS','VISA_CLASS']).size()
```

```
#H-1B_DEPENDENT
LCA_df.groupby(['CASE_STATUS','H-1B_DEPENDENT']).size()
```

```
#WILLFUL_VIOLATOR
LCA_df.groupby(['CASE_STATUS','WILLFUL_VIOLATOR']).size()
```

```
#FULL_TIME_POSITION
LCA_df.groupby(['CASE_STATUS','FULL_TIME_POSITION']).size()
```

```
#NAICS codes
denied_NAICS_codes=np.unique(LCA_df[LCA_df.CASE_STATUS.isin(['Denied'])].NAICS_CODE)
len(denied_NAICS_codes)
```

```
len(np.unique(LCA_df[LCA_df.NAICS_CODE.isin(denied_NAICS_codes) & LCA_df.CASE_STATUS.isin(['Certified'])].NAICS_CODE))
```

```
#wages -- the time scales might not match between the two fields but still only 3 denied here
LCA_df[LCA_df.WAGE_RATE_OF_PAY_FROM.lt(LCA_df.PREVAILING_WAGE)].groupby('CASE_STATUS').size()
```
